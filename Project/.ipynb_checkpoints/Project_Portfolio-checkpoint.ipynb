# Research Portfolio 

### Due April 24, 2020  

The Research Portfolio is a technical writing portfolio blog intended to _teach_ another NEU Master's student a techical subject related machine learning.

It will be the students choice whether to contribute their work to the Introduction to Machine Learning (I2ML) https://compstat-lmu.github.io/lecture_i2ml/index.html and Introduction to Statistical Learning (I2SL) https://github.com/nikbearbrown/I2SL projects.

Look at the course Introduction to Machine Learning (I2ML) https://compstat-lmu.github.io/lecture_i2ml/index.html and then you will propose something to contribute. Contributions can range from making some quiz/test questions, translating some R to python, a lab, a case study or writing an entire additional chapter.

The difference between the I2ML and I2SL projects are the languages the are written in (I2ML is in R and I2SL is in python). I2SL will also have a number of appendices after the theorectectial course which cover the hands-on teaching of essential skills related to data science or more "exotic" algorthms.

For example, MIT has a class called ’The missing semester of your computer science education’ It is a collection of things that most developers and data scientists typically teach themselves on the job such as Setting up projects on Google collab,Shell Tools and Scripting,Data Wrangling, Command-line Environment, Version Control (Git), Debugging and Profiling, Metaprogramming, Security and Cryptography.  The content is available for free at https://missing.csail.mit.edu

Content MUST be written as jupyter notebooks and if you also want to contribute to I2ML then in R as an .Rmd file.  Look at Christoph Molnar's Github https://github.com/christophM/interpretable-ml-book where he uses bookdown and .Rmd files to render the book _Interpretable Machine Learning A Guide for Making Black Box Models Explainable_ https://christophm.github.io/interpretable-ml-book/ and the course Introduction to Machine Learning (I2ML) https://compstat-lmu.github.io/lecture_i2ml/index.html We will do something very similar for I2SL.  Students will contribute to the I2SL and, once accepted, the material will be added to the I2SL course.

_Did I create effective assements? (15 Points)_
You must create assements like quiz questions and programming question that allow a user to self-assess the material you cover.

_Did I explain my idea clearly? (25 Points)_
How effective are you at explaining what you are doing?

_It MUST run and be in a stand alone jupyter notebook. (10 Points)_ 

_What code/writing is yours and what have you adapted? (10 Points)_    
You must explain what code and writing is yours and what you have done that is different. Failure to cite ANY code or writing will result in a zero for this section.

_Did I explain my code clearly? (10 Points)_
Your code review score will be scaled to a range of 0 to 10 and be used for this score.  The code MUST follow Google Python Style Guide http://google.github.io/styleguide/pyguide.html and be in python 3+. 

_Did I explain my licensing clearly? (10 Points)_
Failure to cite a clear license will result in a zero for this section.  This includes whether you wish to contribute to the I2SL project.

_Is the scope enough for a semesters work? (20 Points)_
Is the scope is appropriate for the number of people in the group and it being a semester project?
